1. Pregunta te√≥rica: ¬øQu√© diferencia existe entre la covarianza y la correlaci√≥n?


>- La `covarianza` indica si ambas variables var√≠an en la misma direcci√≥n 
>- **covarianza positiva**: si aumenta el valor de una variable aumenta el variable de la otra 
>- **covarianza negativa**: cuando aumenta el valor de una de las variables, disminuye el de la otra.
>- El valor de la covarianza no es importante, solo el signo (si es positivo o negativo)


>- La `correlaci√≥n` var√≠a entre -1 y 1. Dos variables est√°n asociadas o correlacionadas cuando una variable nos da informaci√≥n acerca de la otra. La correlaci√≥n indica la direcci√≥n (positiva o negativa) y la "fuerza" de esa relaci√≥n.
>
>   - Si el valor de la correlaci√≥n es 0 significa que no hay relaci√≥n entre las variables. 
>
>    - Si el valor es positivo significa que cuando aumenta el valor de una variable aumenta el de la otra. 
>
>    - Si el valor es negativo significa que cuando disminuye el valor de una variable disminuye el de la otra.

3. Con los resultados de la correlaci√≥n del ejercicio anterior, cread un heatmap.


```python
# Creamos una m√°scara
mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))

```

```python
sns.heatmap(df.corr(),
            cmap = "YlGnBu", 
            mask = mask,
           annot = True)
plt.show();
```
